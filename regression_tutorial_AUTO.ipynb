```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
# loading the dataset from excel file (.csv)
from google.colab import files
uploaded = files.upload()
```

```python
# importing the dataset and extracting the columns/data for independent and dependent variables
data = pd.read_csv('tutorial.csv')
x = data.iloc[:, :-1].values
y = data.iloc[:, 3].values
data.head(60)
```

```python
# data visualisation with correlation matrix
sns.heatmap(data.corr())
```

```python
# splitting the dataset into the training set and test set
# x-sensor data (independent variable), y-quality data (dependent) to be predicted
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

```python
# creating and fitting multiple linear regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# predicted variable definition and display of predicted values
# training has 48 and test has 12 data points
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
# multiple linear regression coefficients as in m in the lecture notes
print(regressor.coef_)
```

```python
# y-intercept for the linear equation as in c in the lecture notes
print(regressor.intercept_)
```

```python
# prints y values for test
print(y_test)
```

```python
#same code above
print(y_pred)
```

```python
# calculation of the r-squared value for the regression model (prediction power)
# y_test: given to us as sensor data
# y_pred: predicted results, what we think is true without doing microscopy or microscopy
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
